```python
#Crear un algoritmo para calcular la sumatoria de los primeros cien 
#números (del 01 al 100) con un ciclo while.
     
indice = 0
suma = 0

while indice <= 100:
    suma = suma + indice
    print(suma)
    
    indice += 1
```

```python
# Crea una función llamada verificar_calificacion que reciba tres  parámetros: nota1, nota2 y nota3
#↪ Dentrodela función calcular el promedio de notas
#↪ Si el promedio es mayor o igual a 6, entonces la función debe 
# retornar un mensaje “Aprobado”, en caso contrario
# “Reprobado”

def verificar_calificacion(nota1, nota2, nota3):
  promedio = (nota1 + nota2 + nota3)/3

  if(promedio >= 6):
    return print("Aprobado, tiene un promedio de ", promedio)
  else:
      return print("Desaprobado, tiene un promedio de ", promedio)

verificacion = verificar_calificacion(6, 10, 3)
print(verificacion)

verificacion = verificar_calificacion(5, 4, 3)
print(verificacion)
```
